# 1.2.1: Linear Spaces

In this section, I will:
- Visualize a span of some vector x
- Show that this span is a vector subspace

First, I will initialize all vectors and their respective coordinates


```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# x, y, z components of vector x
x_x = 12
y_x = 0
z_x = -4

# x, y, z components of vector a
x_a = 6
y_a = 3
z_a = -2

# x, y, z components of vector b
x_b = 6
y_b = -3
z_b = -2


x = np.array([x_x, y_x, z_x])
a = np.array([x_a, y_a, z_a])
b = np.array([x_b, y_b, z_b])

```

```python
import numpy as np
import sympy as sy

def make_matrix(np):
  return(sy.Matrix([np[0], np[1], np[2]]))

a_matrix = make_matrix(a)
a_matrix.rref()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D



fig = plt.figure()
subplot3d = plt.axes(projection = '3d')
subplot3d.set_xlim([-12, 12])
subplot3d.set_ylim([-6, 6])
subplot3d.set_zlim([-4,4])


origin = [0,0,0]
subplot3d.quiver(origin[0], origin[1], origin[2], a[0], a[1], a[2], color = 'r')
subplot3d.quiver(origin[0], origin[1], origin[2], b[0], b[1], b[2])
subplot3d.quiver(origin[0], origin[1], origin[2], x[0], x[1], x[2], color = 'g')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from matplotlib.animation import FuncAnimation

fig, ax = plt.subplots(subplot_kw=dict(projection="3d"))

def get_arrow(theta):
    x = np.cos(theta)
    y = np.sin(theta)
    z = 0
    u = np.sin(2*theta)
    v = np.sin(3*theta)
    w = np.cos(3*theta)
    return x,y,z,u,v,w

quiver = ax.quiver(*get_arrow(0))

ax.set_xlim(-2, 2)
ax.set_ylim(-2, 2)
ax.set_zlim(-2, 2)

def update(theta):
    global quiver
    quiver.remove()
    quiver = ax.quiver(*get_arrow(theta))

ani = FuncAnimation(fig, update, frames=np.linspace(0,2*np.pi,200), interval=200)

plt.show()
```
