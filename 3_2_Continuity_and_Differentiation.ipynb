# 3.2.1: Limits and Continuity

In this section, I will prove the following theorum:

Say we have the equation:

$ y = 1/5x^3$ $ + 10x^2 $ $ + 2x $ $ + 21 $

We will show that f(x) becomes closer to f(a) as x also gets closer to a. Here we will use x = 5

```python
from IPython.utils.text import string
# a = 5, f(a) = f(5) = 306

x = 4.0
y = 0.0

# interval = 0.05

print("f(x) as it approaches f(a) from the LEFT\n")
while (x < 5):
  y = 0.20*x**3 + 10*x**2 + 2*x + 21
  print("f(" + str(x) + ") = " + str(y))
  x += 0.05

print("\nf(x) as it approaches f(a) from the RIGHT\n")

x = 6.0
while (x > 5):
  y = 0.20*x**3 + 10*x**2 + 2*x + 21
  print("f(" + str(x) + ") = " + str(y))
  x -= 0.05
```

As you can see, both functions approach f(a) as x becomes closer to a

# 3.3.1: Derivatives

In this section, I will show the following using the previous example as f(x) approaches f(5):

Screenshot 2023-10-06 at 8.29.14 PM.png

We should see that that the limit will produce the slope of f(x)

```python
# a = 5, f(a) = f(5) = 306

x = 5.0
y = 0.0
lim = 0.0
h = -0.02
y_i = 306.0

x = x + h

# interval = 0.01

print("f(x) as it approaches f(a) from the LEFT\n")
while (x < 5):
  y = 0.20*x**3 + 10*x**2 + 2*x + 21
  lim = (y - y_i)/h
  print("lim " + str(x) + " -> 5(-) = " + str(lim))
  x = 5.0
  h += 0.001
  x += h


print("\nf(x) as it approaches f(a) from the RIGHT\n")

x = 5.0
y = 0.0
lim = 0.0
h = 0.02
y_i = 306.0

x = x + h

while (x > 5):
  y = 0.20*x**3 + 10*x**2 + 2*x + 21
  lim = (y - y_i)/h
  print("lim " + str(x) + " -> 5(-) = " + str(lim))
  x = 5.0
  h -= 0.001
  x += h
```

As we can see, the limit gets closer to each other as the left and right side approach 5, zoning into 117
