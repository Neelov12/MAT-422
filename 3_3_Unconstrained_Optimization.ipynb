Let's find P(X = 4)

```python
# Recursive factorial function
def factorial(n):
  if (n <= 1):
    return 1
  else:
    return n*factorial(n-1)

def binomial_n(x_, n_, p_):
  fact = factorial(x_)/(factorial(n_)*factorial(x_ - n_))
  p_n = fact*(p_**n_)*(1. - p_)**(x_ - n_)
  print("P(X = " + str(n_) + ") = ")
  print(p_n)

binomial_n(10, 4, 0.7)
```

Additionally, we can do this for P(X = 1), P(X = 2), ... , P(X = 10)

```python
def binomial(x_, p_):
  f = 0.
  p = 0.
  for n in range(1, x_ + 1):
    f = 0.
    p = 0.
    f = factorial(x_)/(factorial(n)*factorial(x_ - n))
    p = f*(p_**n)*(1 - p_)**(x_ - n)
    print("P(X = " + str(n) + ") = ")
    print("   " + str(p))

binomial(10, 0.7)
```
