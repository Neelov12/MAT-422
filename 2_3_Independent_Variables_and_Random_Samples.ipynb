# 2.3.1: Joint Probability Distributions

In this section, I will demonstrate Joint Probability Distributions by solving the following example:

> Determine the probability that a randomly selected card in a standard 52 deck is 5 and black
>
> And, show that the PMF of all such combinations (e.g. ace and red, ace and black, 2 and red, 2 and black...) sum up to 1.

```python
# Number of cards in deck = 52
# Number of 5's = 4
# Number of blacks = 26

n = 52
x = 4
y = 26

Px = 4/52
Py = 26/52

Px_and_y = Px * Py

print(Px_and_y)
```

Now let's show that
> $\sum_x\sum_yp(x,y) = 1$

```python
# We must find the total number of combinations

# 13 faces, 4 cards per face, 52 total cards, 1/2 chance it is either red of black
com = (13 * 4 * 52)/2

# Now lets find the number of combinations times the p(x,y)

sum = (com * Px_and_y)/52

print(sum)
```

Here we see, the sum of the probability of all combinations is 1

# 2.3.2: Correlation and Dependence

In this section, I will demonstrate correlation and dependence by creating a data set of 50 random numbers from 0-25, and showing that the correlation coefficient is roughly equal to 0

```python
import random
import numpy as np

data_set_x = np.arange(0, 50).reshape(50)
data_set_y = np.arange(0, 50).reshape(50)

for i in range(0, 50):
  data_set_x[i] = random.randint(0, 25)

for i in range(0, 50):
  data_set_y[i] = random.randint(0, 25)

# Determine mean
# of x
sum = 0.0
for i in range(0, 50):
  sum += data_set_x[i]

mean_x = sum/50

# of y
sum = 0.0
for i in range(0, 50):
  sum += data_set_y[i]

mean_y = sum/50


# Calculate covariance
cov = 0.0

for i in range(0, 50):
  cov += (data_set_x[i] - mean_x)*(data_set_y[i] - mean_y)*(1/50)

# Calculate variance of x and y

v_x = 0.0
v_y = 0.0

for i in range(0, 50):
  v_x += (data_set_x[i] - mean_x)*(data_set_x[i] - mean_x)*(1/50)

for i in range(0, 50):
  v_y += (data_set_y[i] - mean_y)*(data_set_y[i] - mean_y)*(1/50)

# Calculate coefficient

p = cov/(v_x*v_y)

print(p)
```

As predicted, the correlation coefficient is roughly 0

# 2.3.3: Random Samples


In this section, I will demonstrate the Central Limit Theorum by plotting distributions from sample sizes small to large to show that the data tilts toward a normal distribution as the sample size increases

```python
import numpy
import matplotlib.pyplot as plt

# Size of sample
sample_size = [1, 25, 100, 1000, 2000, 5000, 10000, 20000, 50000]

# Means of each sample
means = []

for i in sample_size:
    numpy.random.seed(1)
    x = [numpy.mean(numpy.random.randint(-100, 100, i)) for j in range(1000)]
    means.append(x)

# Plot the distributions
fig, ax = plt.subplots(3, 3, figsize =(12, 12))
k = 0
for i in range(0, 3):
    for j in range(0, 3):
        ax[i, j].hist(means[k], 10, density = True)
        ax[i, j].set_title(label = "Sample size =" + str(sample_size[k]))
        k += 1

plt.show()
```
