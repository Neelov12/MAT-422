## 3.5 K Means

I will demonstrate the concept of K-means using Python's built in method

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np

from sklearn.datasets import make_blobs
X, y_true = make_blobs(n_samples=150, centers=5,
                       cluster_std=0.60, random_state=0)
plt.scatter(X[:, 0], X[:, 1], s=50);
```

## 3.6 Support Vector Machine

In this section, I will demonstrate the concept of Support Vector Machine (SVM) using the following example

Imports

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
```

```python
from sklearn.datasets import make_blobs
X, y = make_blobs(n_samples=150, centers=2,
                  random_state=0, cluster_std=0.70)
plt.scatter(X[:, 0], X[:, 1], c=y, s=50, cmap='autumn');
```

As we can see, we have developed two main clusters based on two different population's differing quantitative characteristics. This is especially useful when trying to abstract out differences between two real life populations.

We can also develop a better understanding of these differences by maximizing margins.

```python
xfit = np.linspace(-1, 3.5)
plt.scatter(X[:, 0], X[:, 1], c=y, s=50, cmap='autumn')

for m, b, d in [(1, 0.5, 0.25),
                (0.5, 2.0, 0.33),
                (-0.15, 1.85, 0.3)]:
    yfit = m * xfit + b
    plt.plot(xfit, yfit, '-k')
    plt.fill_between(xfit, yfit - d, yfit + d, edgecolor='none',
                     color='darkseagreen', alpha=0.5)

plt.xlim(-1, 3.5);
```

As we can see, maximizing the margins allow us to develop a better understanding of the qualitative differences between these two datasets
