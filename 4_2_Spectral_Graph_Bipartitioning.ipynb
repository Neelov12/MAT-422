# 4.2 Spectral Graph Bipartitioning

In this section, I will demonstrate the concept of spectral graph bipartitioning by solving the following example:
>
> We would like to show the Laplacian graph of a network consisting of 12 by 12 (144) vertices. Each vertice must have an edge to another and should maximize connectivity while minimizing edge distances.
>
> Use spectral embeddings to do this, and show the resulting network

## Concept

This section will demonstrate the following theorum, where in one dimension:

\begin{equation}
\mathop{\mathsf{minimize}}_x \sum_{(i,j) \in E} (x_i - x_j)^2\\
\text{subject to } x^T 1 = 0, \|x\|_2 = 1
\end{equation}

Note that the objective function is a quadratic form on vector $x$:
\begin{equation}
\sum_{(i,j)\in E} (x_i - x_j)^2 = x^T B B^T x = x^T L x
\end{equation}

Where $B$ is the incidence vector

We can use eigenvectors to determine the next-largest eigenvalues, since vector 1 is the nullspace of $L$, which is equivalent to finding the eigenvector with the second smallest eigenvalue.


### Code Demonstration

Imports

```python
import networkx as nx
import numpy as np
import matplotlib.pyplot as plt
import scipy.linalg as la
import scipy.sparse as sparse
import scipy.sparse.linalg as sla
```

Initialize formulas

```python
G = nx.grid_2d_graph(12,12)
L = nx.laplacian_matrix(G)
L = sparse.csr_matrix(L, dtype=np.float64)
```

Show unconnected graph prior to spectral embedding

```python
lam, V = sla.eigsh(L, which='SM')
plt.scatter(V[:,1], V[:,2])
plt.show()
```

Connected network graph

```python
nx.draw_spectral(G)
```

As we can see, this is the final Laplacian graph
