## Graph

First, let's graph the data points alone

```python
import matplotlib.pyplot as plt
import numpy as np

# 10 total players on each team
y_myteam = np.zeros(10)
y_opteam = np.ones(10)

# Points for each player on each team
x_myteam = np.array([18.67,13.28,14.75,19.9,19.51,13.59,14.72,9.46,9.46,8.68])
x_opteam = np.array([19.51,21.61,15.71,15.99,11.84,7.59,15.73,13.9,7.39,8.13])

x_total = np.concatenate([x_myteam, x_opteam])
y_total = np.concatenate([y_myteam, y_opteam])

plt.scatter(x_myteam, y_myteam)
plt.scatter(x_opteam, y_opteam)
```

Determine slope and intercept of regression line

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression(fit_intercept=True)

model.fit(x_total[:, np.newaxis], y_total)

xfit = np.linspace(0, 25, 1000)
yfit = model.predict(xfit[:, np.newaxis])

plt.scatter(x_total, y_total)
plt.plot(xfit, yfit);

slope = model.coef_[0]
intercept = model.intercept_

print("Model slope:    ", slope)
print("Model intercept:", intercept)
```

Using this equation, I will determine the logistic regression curve

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model.fit(x_total[:, np.newaxis], y_total)

xfit = np.linspace(0, 25, 1000)
yfit = model.predict(xfit[:, np.newaxis])


plt.scatter(x_total, y_total)
plt.plot(xfit, yfit);
```
